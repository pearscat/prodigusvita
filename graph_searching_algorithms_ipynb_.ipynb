```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

def dfs(graph, start):
    visited = []
    stack = [start]
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            visited.append(vertex)
            stack.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)
    return visited

def bfs(graph, start):
    visited = []
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.append(vertex)
            queue.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)
    return visited

def visualize_graph(graph):
    graph_nx = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            graph_nx.add_edge(node, neighbor)
    nx.draw(graph_nx, with_labels=True, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
    plt.show()

# Sample graph
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Test DFS and BFS
dfs_result = dfs(graph, 'A')
bfs_result = bfs(graph, 'A')

# Visualize the graph
visualize_graph(graph)

import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

def dfs(graph, start):
    visited = []
    stack = [start]
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            visited.append(vertex)
            stack.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)
            return visited

def bfs(graph, start):
    visited = []
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.append(vertex)
            queue.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)
    return visited

def visualize_graph(graph):
    graph_nx = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            graph_nx.add_edge(node, neighbor)
    nx.draw(graph_nx, with_labels=True, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
    plt.show()

# Sample graph
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Test DFS and BFS
dfs_result = dfs(graph, 'A')
bfs_result = bfs(graph, 'A')

# Visualize the graph
visualize_graph(graph)

print("DFS traversal:", dfs_result)
print("BFS traversal:", bfs_result)
```
